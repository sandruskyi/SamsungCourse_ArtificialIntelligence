## 2.1 Empezar con OpenCV - Eliger una imagen propia y pasarla a blanco y negro

```python
import cv2
# Se recomienda 
# Generar una variable con el nombre del fichero
name_file = 'shiba.jpg'
# Abrir el fichero con OpenCV
shiba = cv2.imread(name_file)
print(shiba.shape)
# Convertir el fichero a blanco y negro
shibaGray = cv2.cvtColor(shiba,cv2.COLOR_BGR2GRAY)
print(shibaGray.shape)
# Guardar la imagen como IMAGEN_MOD1_tu_nombre.jpg
cv2.imwrite('IMAGEN_MOD1_Sandra_Gomez.jpg', shibaGray)
```

### 2.2 Coger la imagen y pintarla con matplotlib con diferentes mapas de color

```python
import cv2
import matplotlib.pyplot as plt
# Se recomienda 
# Abrir la imagen en blanco y negro con OpenCV
shibaGray = cv2.imread('IMAGEN_MOD1_Sandra_Gomez.jpg',  cv2.IMREAD_GRAYSCALE) 
        #cv2.IMREAD_GRAYSCALE o -1 , si no ponemos nada lo coge como si fuera en color BGR

# Comprobar que la imagen realmente esta en blanco y negro
cv2.imshow('Imagen en blanco y negro', shibaGray)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Investigar los diferentes mapas de color con los que se puede dibujar con matplotlib  (cmap='')

# Generar una figura en matplotlib que contenga varios subplots con los diferentes mapas de color 
fig = plt.figure(figsize=(16,10))

fig.add_subplot(2,3,1)
plt.imshow(shibaGray, cmap='gray')
plt.axis('off')
plt.title("Imagen en b/n")

fig.add_subplot(2,3,2)
plt.imshow(shibaGray, cmap='pink')
plt.axis("off")
plt.title("Imagen en cmap pink")

fig.add_subplot(2,3,3)
plt.imshow(shibaGray, cmap='twilight')
plt.axis("off")
plt.title("Imagen en twilight")

fig.add_subplot(2,3,4)
plt.imshow(shibaGray, cmap ='hsv')
plt.axis("off")
plt.title("Imagen en hsv")

fig.add_subplot(2,3,5)
plt.imshow(shibaGray, cmap='bwr')
plt.axis("off")
plt.title("Imagen en bwr")

fig.add_subplot(2,3,6)
plt.imshow(shibaGray, cmap='PuRd')
plt.axis("off")
plt.title("Imagen en PuRd")

# Presentar la imagen de 6 diferentes formas
plt.show()

## Otra manera de progamarlo en menos l√≠neas:
cmap_types = ['spring', 'prism']

fig = plt.figure(figsize=(16,10))
for i, cmap_type in enumerate(cmap_types):
    fig.add_subplot(1,len(cmap_types)+1,i+1)
    plt.imshow(shibaGray, cmap=cmap_type)
    plt.axis("off")
    plt.title("Imagen en "+ cmap_type)
plt.show()
```

### 3.1 Crear una imagen con tus iniciales o logo de 100x50 pixeles en formato BGR

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Se recomienda: 
# Crear una matriz de numpy con las dimensiones adecuadas
img = np.zeros((100,50,3),dtype=np.uint8) # 100 ancho 50 alto en 3 canales, como est√° rellena de 0s ser√° negra
print(img)
print('Shape', img.shape)

# Identificar el codigo BGR para por lo menos 3 colores que se quieran utilizar. 
# Idealmente se puede generar un diccionario con los colores que se quieren utilizar
misColores = {'pink':(203,192,255), 'violet':(133,21,199), 'deeppink':(147,20,255)}

# Comprobar que la imagen realmente esta en BGR con los colores que se quieren utilizar
#img[:,:]=misColores['pink']
#img[:,:]=misColores['violet']
#img[:,:]=misColores['deeppink']

# Definir los pixeles mediante formulas (a poder ser no uno a uno, ya que hay 5000!)  
img[5:10, 5:45]=misColores['violet']
img[20:30, 10:40]=misColores['pink']
img[30:35, 15:40]=misColores['deeppink']
img[30:45, 10:15]=misColores['pink']
img[45:55, 10:40]=misColores['pink']
img[55:60, 10:35]=misColores['deeppink']
img[55:70, 35:40]=misColores['pink']
img[70:80, 10:40]=misColores['pink']
img[90:95, 5:45]=misColores['violet']

# Vemos si est√° en BGR y lo mostramos con opencv
cv2.imshow('Imagen BGR', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
# Lo vemos en el plot
plt.imshow(img)
plt.axis('off')
plt.title('Imagen sin transformarla a RGB, est√° en BGR')
plt.show()

# Observamos que hay que pasarlo a RGB
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
plt.title('Imagen transformada a RGB')
plt.axis('off')
plt.show()

# Guardar la imagen con el nombre LOGO_tu_nombre.jpg
cv2.imwrite('LOGO_Sandra.jpg',img)
```

### 3.4 Crear una imagen con diversidad de formas, tipos y colores

```python
import numpy as np
import cv2
from random import choice
# Idea basada en la siguiente pagina:
# https://docs.opencv.org/2.4/doc/tutorials/core/random_generator_and_text/random_generator_and_text.html
# Se recomienda:
# Elegir una forma que se quiera utilizar (por ejemplo linea, elipse, ...)
    # ELLIPSE: cv2.ellipse(image, centerCoordinates, axesLength, angle, startAngle, endAngle, color [, thickness[, lineType[, shift]]])
    # LINES : cv2.line(image, start_point, end_point, color, thickness)
    
# Elegir unas dimensiones del lienzo que se quiere utilizar e inicializar el lienzo 
img = np.zeros((1000,1000,3),dtype=np.uint8) # 1000x1000 con 3 canales
# Identificar las propiedades de la forma y todos los posibles valores de las propiedades

# Crear con numpy un espacio random de propiedades para poder

# Hacer un loop para dibujar todas las formas
misColores = {'pink':(203,192,255), 'violet':(133,21,199), 'deeppink':(147,20,255)}
        
for i in range(5):
    color = choice(list(misColores.values()))
    img = cv2.ellipse(img, (np.random.randint(0,1000),np.random.randint(0,1000)), (np.random.randint(0,1000),np.random.randint(0,1000)),0,0,360, color,5)
# Si se quiere hacer algo mas complejo, se pueden combinar formas

# Guardar la imagen como FORMAS_tu_nombre.jpg
cv2.imwrite('FORMAS_Sandra.jpg',img)
```

### 3.5 Incluir en la imagen de 2.2 elementos descriptivos en diferentes colores

```python
import cv2
import matplotlib.pyplot as plt
# Se recomienda:
# Abrir la imagen en blanco y negro
img = cv2.imread('IMAGEN_MOD1_Sandra_Gomez.jpg', cv2.IMREAD_GRAYSCALE)
print(img.shape)
# Convertir la imagen a color 
img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR) 

# Pensar y definir para que sirven las formas en OpenCV
    # Creo que mayormente las utilizaremos para detectar cosas en im√°genes, normalmente con cuadrados o rect√°ngulos.
# Identificar los elementos mas caracteristicos de la imagen con la forma mas adecuada
cv2.rectangle(img, (950,450),(1100,620),(255,0,255), thickness=10, lineType=3, shift=0)
# Incluir texto descriptivo en los elementos caracteristicos 
cv2.putText(img, "Nariz", (950,425), cv2.FONT_HERSHEY_SIMPLEX,0.9, (255,0,255), 3, cv2.LINE_4)


# Muestro: 
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

fig = plt.figure(figsize=(15,10))
plt.imshow(img)
plt.axis('off')
plt.title('Im√°gen con elementos identificados')
plt.show()
```

## BONUS1: Aplica alguno de los ejericicios anteriores a un video

```python
# Al final un video es una secuencia de imagenes. Seguro que algo de lo que has aprendido lo puedes aplicar a un video
import cv2
import numpy as np

file_name = 'Mariposa.mp4'
video = cv2.VideoCapture(file_name)

while(video.isOpened()):
    # Captura frame-by-frame
    ret, frame = video.read()
    if ret == True: # Si est√° abierto
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) # De bgr a gray
        frame = cv2.applyColorMap(frame, cv2.COLORMAP_JET) # Con colorines xD
        
        cv2.imshow('Frame',frame)
        
    if cv2.waitKey(25) & 0xFF == ord('q'):
        break
    #break the loop
    else: 
        break
video.release()
cv2.destroyAllWindows()
        
```
