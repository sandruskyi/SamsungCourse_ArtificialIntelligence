## 2. Empezamos con OpenCV 
### 2.1 Leer, cambiar formato y mostrar imagenes
Empezaremos leyendo, cambiando el formato y mostrando imagenes con OpenCV

```python
# Importamos la libreria OpenCV
import cv2

# Identificamos el nombre de la imagen que esta en la carpeta del jupyter notebook
file_name = '01_FalconHeavy1-980x724.jpg'

# Leemos la imagen con nombre file_name, en este caso en formato jpg
img = cv2.imread(file_name)

print(img.shape) #Así es que está leyendo bien la imágen

# Convertimos la imagen a blanco y negro
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # cvtColor coge una imagen y la transforma

# Mostramos ambas imagenes con OpenCV
cv2.namedWindow('Imagen en color', cv2.WINDOW_NORMAL)
cv2.imshow('Imagen en color',img) # Lo abre en ventana
cv2.imshow('Imagen blanco y negro',gray)

# Esperamos que se pulse una tecla para cerrar las ventanas
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Guardamos con imwrite la imagen en el nuevo path
cv2.imwrite(file_name.split(".jpg")[0]+"_BYN.jpg", gray) # La guarda en gris
```

### 3.2 Crear imagenes y modificar pixeles
Vamos a crear una imagen aleatoria para entender la estructura de una imagen

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Creamos una imagen aleatoria con numpy
new_image = np.random.randint(low=0, high=255, size=(6,8,3), dtype=np.uint8)

# Mostramos la imagen aleatoria
plt.imshow(new_image)
plt.axis("off")
plt.title('Imagen aleatoria')

plt.show()

# Mostramos que hay detras de la imagen
print("Esto es la imagen\n\nShape: ", new_image.shape, "\n[\n")
for i in range(3):
    print(new_image[:,:,i] , "\n")
print(" ]")
```

### 3.3 Modificar el formato de las imagenes
Vamos a convertir la imagen a diferentes formatos y ver lo que pasa

```python
# Como hemos visto nuestra imagen es RGB, que es como matplotlib pyplot interpreta las imagenes de color. 

# Convertimos la imagen a blanco y negro
new_image_BGR = cv2.cvtColor(new_image, cv2.COLOR_RGB2BGR)
# Convertimos la imagen a blanco y negro
new_image_HSV = cv2.cvtColor(new_image, cv2.COLOR_RGB2HSV)
# Convertimos la imagen a blanco y negro
new_image_GRAY = cv2.cvtColor(new_image, cv2.COLOR_RGB2GRAY)

print("Converter RGB to BGR index: ", cv2.COLOR_RGB2BGR) # Al convertir le añade esos nºs que llamamos index 
print("Converter RGB to HSV index: ", cv2.COLOR_RGB2HSV)
print("Converter RGB to BGR index: ", cv2.COLOR_RGB2GRAY)

print("\nExisten muchos conversores diferentes en OpenCV\n\nAqui esta la lista completa:\n")
flags = [i for i in dir(cv2) if i.startswith('COLOR_')]
print( flags )
```

```python
# Vamos a definir unas cuantas funciones auxiliares para dibujar las comparativas de las imagenes transformadas

def plot_image(image_RGB, image, title=""):
    # Funcion para mostrar una imagen
    # Inputs: 
    # image_RGB - matrix de numpy de referencia con la imagen en RGB
    # image - matriz de numpy con la imagen en otro formato que se mostrara como si fuera RGB
    # title - string con el titulo que saldra en la imagen
    # Outputs:
    # show - de la imagen en cuestion
    from matplotlib import pyplot as plt
    
    # Creamos una figura para hacer los subplots en ella
    fig=plt.figure(figsize=(16, 10))
    
    # Dibujamos la primera imagen - en color, pero en BGR - en la posicion 1 de la figura de forma (2,2)
    fig.add_subplot(1,2,1)
    plt.imshow(image_RGB)
    plt.axis("off")
    plt.title('Imagen referencia - RGB')

    # Mostramos la imagen
    fig.add_subplot(1,2,2)
    plt.imshow(image)
    plt.axis("off")
    plt.title('Imagen convertida - '+title)
    
    plt.show()
    
def plot_matrix(image, title=""):
    # Mostramos que hay detras de la imagen
    print("Esta es la matrix detras de la imagen:",title,"\n\nShape: ", image.shape, "\n[\n")
    
    if len(image.shape) == 3:
        for i in range(3):
            print(image[:,:,i] , "\n")
        print(" ]")
    else: 
        print(image)
```

```python
# Ya que hemos definido nuestra funcion, pues mostramos la informacion de los 4 formatos
plot_image(new_image, new_image, "RGB")
plot_matrix(new_image, "RGB")
```

```python
# Mostramos la imagen transformada a HSV
plot_image(new_image, new_image_HSV, "HSV")
plot_matrix(new_image_HSV, "HSV")
```

### 3.4 Modificar la imagen introduciendo formas con OpenCV
OpenCV te deja incluir formas y texto en las imagenes, para poder mostrar los resultados sobre las imagenes.

```python
import cv2
from matplotlib import pyplot as plt

# Image taken from https://www.emirates.com/media-centre/emirates-to-operate-a380-to-glasgow/
image_A380 = cv2.imread('02_1920_A380-339494_EMIRATES.jpg')
image_A380_RGB = cv2.cvtColor(image_A380, cv2.COLOR_BGR2RGB)
fig=plt.figure(figsize=(18, 16))
plt.imshow(image_A380_RGB)
plt.axis("off")
plt.title('Emirates A380 Glasgow')

print("Emirate Airbus A380-800 \nImage shape is: ", image_A380.shape)
plt.show()
```

```python
# Se pueden dibujar lineas
# cv2.line(img, pt1, pt2, color, thickness=1, lineType=8, shift=0)
image_A380 = cv2.line(image_A380, (0,1200),(1920,1200),(0,0,0), thickness=5, lineType=8, shift=0)

# Se pueden dibujar circulos
# cv2.line(img, pt1, pt2, color, thickness=1, lineType=8, shift=0)
image_A380 = cv2.circle(image_A380, (50,1250),20,(255,255,0), thickness=5, lineType=3, shift=0)

# Se pueden dibujar rectangulos con texto debajo
# cv2.line(img, pt1, pt2, color, thickness=1, lineType=8, shift=0)
image_A380 = cv2.rectangle(image_A380, (500,690),(670,810),(255,0,255), thickness=10, lineType=3, shift=0)
image_A380 = cv2.putText(image_A380, "Motor", (500,850),cv2.FONT_HERSHEY_SIMPLEX, 0.9,
                         (255,0,255), 3 , cv2.LINE_4)

# Y texto incorporando la informacion que haga falta:
from datetime import datetime
import os

image_A380 = cv2.putText(image_A380, "Curso Samsung - Big Data & Artificial Intelligence " 
                + str(datetime.now().strftime("%d-%m-%Y %H:%M:%S")), (200,1250), 
                cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255,255,0), 3 , cv2.LINE_4)

image_A380 = cv2.putText(image_A380, ("Carpeta de trabajo: " + str(os.getcwd())), (50,30), 
                cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255,255,0), 3 , cv2.LINE_4)

image_A380_RGB = cv2.cvtColor(image_A380, cv2.COLOR_BGR2RGB)
fig=plt.figure(figsize=(18, 16))
plt.imshow(image_A380_RGB)
plt.show()
```
