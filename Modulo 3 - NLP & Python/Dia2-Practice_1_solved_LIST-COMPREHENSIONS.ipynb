```
"""DATA"""
# (Los textos t1 y t2 eran para otro ejercicio, se han elminado)

t3 = 'maaaa' 
t4 = 'aubabababbb'
t5 = 'abyababcbcbib'
t6 = 'aba_babe4'
t7 = 'basdb98709e7'
t8 = 'peridise_'
t9 = 'genialees'
t10 = 'xxx333nnn'
t11 = 'braba..n!'

t12 = "Estudios Xrecientes comprueban Xque uno puede irseX y Xvolver de cualquier lugar Xsin publicarlo"
```

```
"""1.Imprime todos los elementos (caracteres) de t4 que sean iguales a 'a'. 
Asume que no conoces los contenidos.  
    a.) Hazlo 1º con un sencillo for loop - así podéis comparar las dos estructuras""" 
    
elems = []
for elem in t4: 
    if elem == 'a': 
        elems.append(elem)
print(elems)
```

```
"""1.b) Hazlo con comprehension list: """

# El cambio de un for in 'vertical' a un for in 'horizontal'  es sencillo: 

elems = [elem for elem in t4 if elem == 'a']

print(elems)
```

```
""" 3. Imprime todos las ocurrencias de'ab' en t4 """

# Ahora no podemos recurrir a una iteración simple, porque se itera sobre  
# las unidades del 'iterable' (los más típicos iterables son las secuencias, 
# los tipos estándares de secuencias son las cadenas, listas o tuples)
# Y el elemento de una cadeana es UN caracter, con lo cual tendríamos que 
# comprobar también el siguiente caracter:  

print(['ab' for i, char in enumerate(t4) if char == 'a' and t4[i+1] == 'b'])    

# Pero queda un problema por resolver, lo veis? (tiene que ver con los índices)

# No leais la siguiente línea si podéis resolverlo, porque es otra pista

# Pista 2. qué pasa en t4[i+i]
```

```
# El proplema anterior es que iteramos sobre i, con lo cual si llegáramos al final, 
# obligaríamos al código a buscar un índice que no existe (el utltimo + 1)
# Si conocéis el método de 'startwith() para cadenas, podemos evitar ese 
# problema con una estructura de loop que habéis seguido bastantes 
# (me refiero al i for i, y al t4[idx:idx+2], no al .startwith()):

substring = 'ab'
res = [i for i in range(len(t4)) if t4.startswith(substring, i)] 

# Devolviendo i, necesitaríamos hacer esto que habes hecho muchas:
print([t4[idx:idx+2] for idx in res]) 
```

```
# El primer elemento de la comp. list, en el caso arriba la i, es el que
# se devuelve para construir la lista. 
# Pero no es necesario devolver lo mismo sobre lo que iteramos! 
# Podemos símplemente seleccionar 'ab' como el elemento que se va a incluir en 
# la línea, en vez de poner el índice: 

['ab' for i in range(len(t4)) if t4.startswith('ab', i)] 
```

```
# Ahora, en este caso, si conocéis el método .count() (agradecemientos a Marga), 
# que funciona tanto en cadenas como listas, en un caso real no haría falta usar
# una comprehension list, salvo por separar las ocurrencias. 
# Tampoco sería necesario recurrir a la c. list para separarlas, pero intentar 
# usarla igual como ejercicio para que podáis aplicarla cuando no conozcáis 
# métodos específicos. Pero ya que conocéis este, lo hace hiper sencillo y aquí 
# no habría diferencia en tiempo de ejecución:

n = t4.count('ab')
print('ab'* n)  # Como tenemos el número de 'ab's, símplemente se imprimen 
```

```
# En realidad podríamos usar una comprenhension list tb con .count. Así podéis 
# manejar cada ocurrencia por separado en vez de tener todo en una string, no
# no necesitaríamos usar el condicional de la c. list, lo que nos viene muy bien
# también para que recordéis que es optativo y quedaría muy elegante: 

['ab' for i in range(t4.count('ab'))]
```

```
# Si no conocéis .startswith() ni .count(), aún no estáis acostumbradas 
# a mirar la documentación y/o no podíais con la compr. list, 
# también podías haberlo hecho con un simple while que sí sabeis manejar. Pero 
# este ejemplo está puesto porque la mayoría habéis recurrido a for loops más 
# complejos, y también para que veáis que prác hacer 
# con pocas instrucciones. Estas instrucciones las vais a tener que usar en 
# cualquier caso por muchas otras que vayáis aprendiendo, y os van a venir muy 
# bien, pero si hebéis tirado por aqui, intendad rehacerlo ahora vosotras con
# una comprehension list:  

while 'ab' in t4: 
  print('ab')
  ab_index = t4.index('ab')  # ab_index coge el índice donde ocurre 'ab' por 1ª vez
  t4 = t4[ab_index+2:]  # quitamos de la string hasta donde aparece este primer 'ab' incluído. 
```

**Función de búsqueda de las ocurrencias**, a la que podemos llamar después:

```
# Versión generalizada de la compr. list para cualquier cadena y subcadena  
# (nos ahorra reescribirlo en otros ejercicios que piden lo mismo o parecido):

def ocurrencias(t, substring):  
    """
    t: string
    substring: string que es parte de la string anterior
    """
    all_occ = [substring for i in range(len(t)) if t.startswith(substring, i)]
    return all_occ
print(ocurrencias(t4, 'ab'))
```

```
""" 5. Imprime todas las ocurrencias de 'ab' en los textos t3 a t11 
Almacena todos los textos primero en la variable alma_1 y recorrelos """

alma_1 = [t3, t4, t5, t6, t7, t8, t9, t10, t11]

print([ocurrencias(string, 'ab') for string in alma_1])
```

```
"""7. Concatena los textos en alma_1 (ej 5) e imprime todas las ocurrencias 
de 'aba' en el texto resultante"""

# 1º. Cogemos la string concatenando todas:
new_st = ''
for st in alma_1:  # no podemos usar una comp_list si no queremos list sino str
    new_st += st
    # print(new_st)  # Imprime new string si quieres ver los valores q toma
    
# 2º. Imprimimos las veces que ocurre 'aba', aprovechando la fn q hemos hecho: 
print(ocurrencias(new_st, 'aba'))
```

```
# 1º Podemos usar una comprh. list + join en vez de concatenar: 
ocurr = [st for st in alma_1]  # Pasaríamos primero todo a una lista
new_st = ''.join(ocurr)  # Y lo unimos con un join

# 2º Como ya tenemos la string (cadena), podemos hacer igual q hemos hecho antes: 
print(ocurrencias(new_st, 'aba'))
```

```
# Por si acaso, notar que el código de la celda anterior es = a esto: 

new_st = ''.join([st for st in alma_1])  # 1º
print(ocurrencias(new_st, 'aba'))  # 2º
```

```
# E igual a esto (sólo que hacemos todo en un paso): 

print(ocurrencias(''.join([st for st in alma_1]), 'aba'))
```

```
""" 9. Almacena en una variable resul_1 el penúltimo caracter de cada string 
si el primer caracter de la string no es una 'm'. Imprime el resultado."""  # Muchas lo tienen 

resul_1 = [string[-2] for string in alma_1 if string[0] != 'm']
print(resul_1)
```

```
"""10. Itera de t3 a t11 y devuelve el primer caracter de la primera string, \
 el segundo de la segunda, el tercero de la tercera... etc. 
 Guarda cada caracter en resul_2 de forma que quede una string con cada uno
 de los caracteres por orden"""

resul_2l = []
for i, t in enumerate(alma_1): 
    for j, char in enumerate(t):
        # Como i es el índice de las strings, y j el de los caracteres, ...
        # ... cuando estos son iguales, añadimos el caracter a la lista
        if i == j:  
            resul_2l.append(char)
            resul_2 = ''.join(resul_2l)
print(resul_2)
```

```
# Esto nos permite ver la utilidad de las comp. lists anidadas (más allá de que sean más rápidas): 

print(''.join([char for i, t in enumerate(alma_1) for j, char in enumerate(t) if i==j]))
```
