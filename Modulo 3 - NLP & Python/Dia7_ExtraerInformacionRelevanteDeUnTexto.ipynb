```

text = "London is the capital and largest city of England and the United Kingdom. Standing on the River " \
           "Thames in the south-east of England, at the head of its 50-mile (80 km) estuary leading to " \
           "the North Sea, London has been a major settlement for two millennia. " \
           "Londinium was founded by the Romans. The City of London, " \
           "London's ancient core − an area of just 1.12 square miles (2.9 km2) and colloquially known as " \
           "the Square Mile − retains boundaries that follow closely its medieval limits." \
           "The City of Westminster is also an Inner London borough holding city status. " \
           "Greater London is governed by the Mayor of London and the London Assembly." \
           "London is located in the southeast of England." \
           "Westminster is located in London." \
           "London is the biggest city in Britain. London has a population of 7,172,036."

# SACAR INFORMACIÓN DEL TEXTO
# Primero buscar sentencias, es decir, separar el texto
# Segundo, creamos el árbol de cada sentencia (Displacy)
# A partir de ahí ir buscando triadas, es decir ir buscando información
# Podemos buscar la raiz y lo que esté al lado nos proporciona información, para ello utilizar patrones


# Ejericio 2: Encontrar todas las ciudades que aparecen en el texto
```

```
import spacy
from spacy import displacy
from spacy.matcher import Matcher


#spacy.cli.download('en_core_web_sm')
nlp = spacy.load('en_core_web_sm')
doc = nlp(text)

roots = []
for sentence in doc.sents: 
  print(sentence)
  displacy.render(sentence, style="dep", jupyter=True) # Mostramos el árbol de cada sentencia
  roots.append(sentence.root) # Añadimos las raices
displacy.render(doc, style="dep", jupyter=True) # Mostramos el árbol de todo el doc

# Teniendo ya las raices, podemos crear un patrón: 
matcher = Matcher(nlp.vocab)
for root in roots: 
  print('Patron')
  # Un patron suele ser una triada
  print('Root',root)
  pattern = [{'LEMMA':root, 'POS':'VERB'}, # Un token
             {'POS':'NOUN'}] #Otro token
  matcher.add("VERBOS_PATTERN", None, pattern)

  matches = matcher(doc)
  for match_id, start, end in matches: 
    matched_span = doc[start:end]
    print(matched_span.text)
```
